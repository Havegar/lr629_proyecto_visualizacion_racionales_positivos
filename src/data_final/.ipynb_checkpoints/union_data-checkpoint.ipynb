```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark import SparkContext, SparkConf
import pyspark
import pandas as pd
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
from pyspark.sql.types import StringType,TimestampType
import matplotlib.pyplot as plt
```

```python
ss_name = 'Agregación de labels a contaminados'
wg_conn = "spark.kerberos.access.hadoopFileSystems"
db_conn = "abfs://data@datalakesii.dfs.core.windows.net/"

spark = SparkSession.builder \
      .appName(f"Ejecucion algoritmo {ss_name}")  \
      .config(wg_conn, db_conn) \
      .config("spark.executor.memory", "6g") \
      .config("spark.driver.memory", "12g")\
      .config("spark.executor.cores", "4") \
      .config("spark.executor.instances", "5") \
      .config("spark.driver.maxResultSize", "12g") \
      .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

spark.conf.set("spark.sql.parquet.enableVectorizedReader","false")
spark.conf.set("spark.sql.parquet.int96RebaseModeInRead", "CORRECTED")
spark.conf.set("spark.sql.parquet.int96RebaseModeInWrite", "CORRECTED")
spark.conf.set("spark.sql.parquet.datetimeRebaseModeInRead", "CORRECTED")
spark.conf.set("spark.sql.parquet.datetimeRebaseModeInWrite", "CORRECTED")
```

```python
df_spark = spark.read.parquet("abfs://data@datalakesii.dfs.core.windows.net/DatosOrigen/lr-629/propuesta_f29_comunidades/data_contaminados_with_labels")
```

```python
df_pandas = df_spark.toPandas()
spark.stop()
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Función para actualizar el gráfico basado en los inputs del usuario
def update_graph(region):
    filtered_df = df_pandas[df_pandas['unidad_regional'] == region]
    plt.figure(figsize=(10, 6))
    plt.hist(filtered_df['score'], bins=20, alpha=0.7)
    plt.title(f'Score Distribution in {region}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.show()

# Crear un widget de selección para las regiones
regions = df_pandas['unidad_regional'].unique()
region_selector = widgets.Dropdown(
    options=regions,
    description='Region:',
)

# Enlazar el widget al gráfico
widgets.interactive(update_graph, region=region_selector)
```

```python
import gradio as gr
import matplotlib.pyplot as plt
import io
import numpy as np
import pandas as pd


def plot_score_distribution(region):
    filtered_df = df_pandas[df_pandas['unidad_regional'] == region]
    plt.figure(figsize=(10, 6))
    plt.hist(filtered_df['score'], bins=20, alpha=0.7)
    plt.title(f'Score Distribution in {region}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')

    # Save plot to a BytesIO object
    img_bytes = io.BytesIO()
    plt.savefig(img_bytes, format='png')
    img_bytes.seek(0)
    plt.close()
    
    return img_bytes.getvalue()

def main():
    regions = df_pandas['unidad_regional'].unique().tolist()  # Convertir a lista
    # Asegurarse de que todas las opciones sean cadenas
    regions = [str(region) for region in regions]
    interface = gr.Interface(
        fn=plot_score_distribution,
        inputs=gr.Dropdown(choices=regions, label="Select Region"),
        outputs=gr.Image(type="bytes")  # Cambiado a "bytes"
    )
    interface.launch(share=True)

if __name__ == "__main__":
    main()
```
